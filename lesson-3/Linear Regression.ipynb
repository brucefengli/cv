```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt



def draw_points(x_list, y_list):
    plt.scatter(x_list, y_list)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()    

'''
draw image
'''   
def draw_line(w, b, line_color='#0000FF'):
    #plt.scatter(x_list, y_list)
    #plt.xlabel('X')
    #plt.ylabel('Y')
    print('w:{0}, b:{1}'.format(w, b))
    plt.plot([0, 100],[0, 100 * w + b], color=line_color)
    

def draw_result():   
    plt.show()
    

'''
evaluate loss
'''
def eval_loss(w, b, x_list, gt_y_list):
    avg_loss = 0.0
    for i in range(len(x_list)):
        avg_loss += 0.5 * (w * x_list[i] + b - gt_y_list[i]) ** 2   #cost function
    avg_loss /= len(gt_y_list)   #cost function
    return avg_loss

'''
inference by w * x + b
'''
def inference(w, b, x):
    pred_y = w * x + b
    return pred_y


'''
evaluate loss
'''
def get_gradient(pred_y, gt_y, x):
    diff = pred_y - gt_y
    dw = diff * x    #get gradient by di
    db = diff
    return dw, db


def cal_gradient(batch_x_list, batch_gt_y_list, w, b, lr):
    avg_dw, avg_db = 0, 0
    batch_size = len(batch_x_list)
    
    for i in range(batch_size):
        pred_y = inference(w, b, batch_x_list[i])
        dw, db = get_gradient(pred_y, batch_gt_y_list[i], batch_x_list[i])
        avg_dw += dw    #cal w differential 
        avg_db += db    #get b differential
        
    avg_dw /= batch_size   #cost function avg by m
    avg_db /= batch_size   #cost function avg by m
    w -= lr * avg_dw    #move along gradient direction
    b -= lr * avg_db    #move along gradient direction
    return w, b
  

'''
train
'''     
def train(x_list, gt_y_list, batch_size, lr, max_iter):    
    w = 0
    b = 0
    list_num = len(x_list)
    for i in range(max_iter):
        batch_idxs = np.random.choice(len(x_list), batch_size)
        batch_x = [x_list[j] for j in batch_idxs]
        batch_y = [gt_y_list[j] for j in batch_idxs]
        w, b = cal_gradient(batch_x, batch_y, w, b, lr)
        print('i: {0}, w:{1}, b:{2}'.format(i, w, b))
        print('loss is {0}'.format(eval_loss(w, b, x_list, gt_y_list)))
        draw_line(w, b)

    print('w:{0}, b{1}'.format(w, b))
    print('loss is {0}'.format(eval_loss(w, b, x_list, gt_y_list)))        
    draw_line(w, b, '#FF0000')

    
'''
generate data
'''
def gen_data():
    #w = random.randint(0, 100) + random.random()
    #b = random.randint(0, 50) + random.random()
    
    w = random.randint(0, 100) + random.random()  # for noise random.random[0, 1)
    b = random.randint(0, 50) + random.random()
    
    num_samples = 1000
    
    #x_list = random.uniform(0, 100, 1000)
    x_list = []
    y_list = []
    
    for i in range(num_samples):
        x = random.randint(0, 100) * random.random()
        y = w * x + b + random.random() * random.randint(-10, 10)
        x_list.append(x)
        y_list.append(y)    

    return x_list, y_list, w, b   


    
'''
linear regression
'''    
def run_linear_regression():
    lr = 0.0001
    max_iter = 1000
    batch_size = 80
    x_list, y_list, w, b = gen_data()
    draw_points(x_list, y_list)
    train(x_list, y_list, batch_size, lr, max_iter)



run_linear_regression()


    
```
